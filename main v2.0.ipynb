```python
gutpath = os.path.expanduser("~/nltk_data/corpora/gutenberg/")
print(gutpath)
```

```python
moby_path = (gutpath+"hom1.txt")
print(moby_path)
```

This METHOD puts all the book in lower case and splits them up so they are not all one string.


```python
import pprint #pretty print
with open(moby_path, "r", encoding="utf-8") as file:
#with open("/home/pi/nltk_data/corpora/gutenberg/hom1.txt", "r", encoding="utf-8") as file:
    text = file.read().lower()
    text_clean = []
    for ch in text:
        if ch not in punctuation:
            text_clean.append(ch) #there are faster but less clear ways to do this.
            #no more punctuation
    tokens = "".join(text_clean).split(" ")
    print(tokens)
    
```

```python
moby = tokens
```

```python
print("our sample has this many words: ", len(moby)) #len is an example of polymorphic code because it applies on various different things.
count = {}
for word in moby:
    count.setdefault(word, 0)
    count[word] += 12
pprint.pprint(count)
    
```

```python
for word in sorted(count, key=count.get, reverse=True):
    print(word, count[word])
```
